```python
'''
Created on 2026. 1. 15.

@author: chaliepark
'''


class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            self.rect()

        elif self.square == 2:
            self.par()

        elif self.square == 3:
            self.trape()

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        self.result = '직사각형의 넓이는 : ' + str(area)
        return self.result

    def par(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        self.result = '평행사변형의 넓이는 : ' + str(area)
        return self.result

    # 사다리꼴 메서드
    def trape(self):
        top_width,bottom_width, vertical = map(int, input('윗변,밑변,높이를 입력하세요. 예시 : 윗변,밑변,높이\n >>>').split(','))
        area = (top_width+bottom_width) *vertical/ 2
        self.result = '사다리꼴의 넓이는 : ' + str(area)
        return self.result

b = Square()  # 객체 생성 & 2, 3을 각각 입력해 봅시다.

print(b.result)

```
